# Simple waveguide study
In this notebook we use the libraries in `waveguide_library.py` <br>

```python
import numpy as np
import matplotlib.pyplot as plt
from qdast.modeling.utils.resonators import cpw_capacitance_inductance
from qdast.modeling.waveguides.libraries.waveguide_library import waveguide_library
```

```python
waveguide_lib = waveguide_library()
waveguide_lengths = np.linspace(10, 500, 1001)

# Analytical expectations (ignoring boundary effects):
expected_C11 = cpw_capacitance_inductance(10e-6, 6e-6, waveguide_lengths*1e-6, 11.9)[0]
C11 = waveguide_lib(waveguide_lengths)
```

```python
plt.plot(waveguide_lengths, C11)
plt.plot(waveguide_lengths, expected_C11)

plt.xlabel("Waveguide length (um)")
plt.ylabel("Capacitance to ground")
plt.legend(["Analytical", "Ansys Q3D"])
plt.show()
```
