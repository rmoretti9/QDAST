# 3-ports Clockmon library analysis

```python
import numpy as np
import matplotlib.pyplot as plt
from modeling.qubits.clockmon.libraries.clockmon_library_3ports import clockmon_library_3ports, get_csigma_cqr

# With 3 ports, qucat find spurious moded that generate warnings. Disable the bottom lines to see the warnings.
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
clockmon_lib = clockmon_library_3ports(deembed = 200, port_id = "3_4_5")
coupler_widths3 = np.linspace(30, 200, 101)
coupler_widths4 = 50
coupler_widths5 = 80
```

```python
Cmat = clockmon_lib(coupler_widths3, coupler_widths4, coupler_widths5)
```

```python
c_s = np.zeros(Cmat.shape[0])
c_qr_3 = np.zeros(Cmat.shape[0])
c_qr_4 = np.zeros(Cmat.shape[0])
c_qr_5 = np.zeros(Cmat.shape[0])
for i in range(Cmat.shape[0]):
    c_s[i], c_qr_3[i], c_qr_4[i], c_qr_5[i] = get_csigma_cqr(Cmat[i])
```

```python
plt.plot(c_s)
```

```python
plt.plot(coupler_widths3, c_qr_3)
plt.plot(coupler_widths3, c_qr_4)
```
